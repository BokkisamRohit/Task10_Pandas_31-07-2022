15 . try to find out the 10th most loss making product catagory .

```python
#Solution for q15
df.groupby(['Product Category'])['Profit'].apply(lambda x: x[x<0].sum()).reset_index()
```

Since there are only three product categories we can't find the 10th loss making product category. Therefore in following code cells I will be using both product category and product sub-category to groupby and find out 10th most loss making Product Category.

```python
df15=df.groupby(['Product Category','Product Sub-Category'])['Profit'].apply(lambda x: x[x<0].sum()).reset_index()
df15_1=df15.sort_values(by=['Profit'])
df15_1
```

```python
df15_1.iloc[9:10]
```

The 10th most loss making product category is Technology under which Computer Peripherals is 10th most loss making Product Sub-Category with overall loss of -56544.485587

16 . Try to find out 10 top  product with highest margins
